```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_pickle('../data/final_features.pkl')
```

```python
df.drop(columns=['CustomerID','LoanApplicationDate'],inplace=True)
df = df.set_index('LoanId')
```

```python
df.head()
```

```python
Train_set, Test_set, _, _ = train_test_split(
    df,df['Target'],
    test_size=0.2,         # 20% goes to test set
    random_state=42,       # for reproducibility
    stratify=df['Target']  # ensures the same target distribution in train/test
)
```

```python
Train_set.to_pickle('../data/train_set.pkl')
Test_set.to_pickle('../data/test_set.pkl')
```
